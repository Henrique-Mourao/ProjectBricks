```python
# Notebook: clean_and_transform
# Descrição: Limpa e transforma dados para camada Silver

from pyspark.sql.functions import col, to_timestamp

bronze = spark.table("weather_bronze")

print(f"Lendo {bronze.count()} registros da Bronze")

silver = bronze.select(
    col("city"),
    col("latitude").cast("double"),
    col("longitude").cast("double"),
    col("temperature").cast("double"),
    col("windspeed").cast("double"),
    col("winddirection").cast("integer"),
    col("weathercode").cast("integer"),
    to_timestamp("time").alias("observation_time"),
    to_timestamp("ingestion_timestamp").alias("ingestion_time")
)

silver.write.format("delta") \
  .mode("overwrite") \
  .option("mergeSchema", "true") \
  .saveAsTable("weather_silver")

print(" Dados salvos na tabela weather_silver")
print(f" Total de registros na Silver: {silver.count()}")

display(silver.limit(10))
```
