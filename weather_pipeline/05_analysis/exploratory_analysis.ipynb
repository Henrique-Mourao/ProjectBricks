```python
# Notebook: exploratory_analysis
# Descrição: Análise exploratória dos dados

from pyspark.sql.functions import col

print("\n ANÁLISE DE TEMPERATURA")

gold_temp = spark.table("weather_gold_temperature")

print("\n Estatísticas Gerais:")
gold_temp.describe().show()

print("\n TOP 5 CIDADES MAIS QUENTES:")
gold_temp.orderBy(col("avg_temperature").desc()).limit(5).show(truncate=False)

print("\n TOP 5 CIDADES MAIS FRIAS:")
gold_temp.orderBy(col("avg_temperature").asc()).limit(5).show(truncate=False)

print("\n ANÁLISE DE VENTO")

gold_wind = spark.table("weather_gold_wind")

print("\n TOP 5 CIDADES MAIS VENTOSAS:")
gold_wind.orderBy(col("avg_windspeed").desc()).limit(5).show(truncate=False)

print("\n RESUMO GERAL")

summary = spark.sql("""
    SELECT 
        COUNT(*) as total_cidades,
        ROUND(AVG(avg_temperature), 2) as temp_media_global,
        ROUND(MAX(max_temperature), 2) as temp_mais_alta,
        ROUND(MIN(min_temperature), 2) as temp_mais_baixa
    FROM weather_gold_temperature
""")

summary.show(truncate=False)

print("\n Análise exploratória concluída!")
```
