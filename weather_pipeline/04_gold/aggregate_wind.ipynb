```python
# Notebook: aggregate_wind
# Descrição: Agrega dados de vento

from pyspark.sql.functions import avg, max, min, round as spark_round, col

silver = spark.table("weather_silver")

print(f" Lendo {silver.count()} registros da Silver")

gold_wind = silver.groupBy("city").agg(
    spark_round(avg("windspeed"), 2).alias("avg_windspeed"),
    spark_round(max("windspeed"), 2).alias("max_windspeed"),
    spark_round(min("windspeed"), 2).alias("min_windspeed")
).orderBy(col("avg_windspeed").desc())

gold_wind.write.format("delta") \
  .mode("overwrite") \
  .option("mergeSchema", "true") \
  .saveAsTable("weather_gold_wind")

print(" Tabela weather_gold_wind criada")
print(f" Total de cidades: {gold_wind.count()}")

display(gold_wind)
```
