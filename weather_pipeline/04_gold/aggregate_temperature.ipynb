```python
# Notebook: aggregate_temperature
# Descrição: Agrega dados de temperatura para análise

from pyspark.sql.functions import avg, max, min, count, round as spark_round, col

silver = spark.table("weather_silver")

print(f" Lendo {silver.count()} registros da Silver")

gold_temp = silver.groupBy("city").agg(
    spark_round(avg("temperature"), 2).alias("avg_temperature"),
    spark_round(max("temperature"), 2).alias("max_temperature"),
    spark_round(min("temperature"), 2).alias("min_temperature"),
    count("*").alias("total_observations")
).orderBy(col("avg_temperature").desc())


gold_temp.write.format("delta") \
  .mode("overwrite") \
  .option("mergeSchema", "true") \
  .saveAsTable("weather_gold_temperature")

print("Tabela weather_gold_temperature criada")
print(f"Total de cidades: {gold_temp.count()}")

display(gold_temp)
```
