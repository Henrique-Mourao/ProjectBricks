```python
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import os
import pandas as pd
from datetime import datetime

#Passivo monetário do Banco Central 
caminho = "/Workspace/Users/henriquegmour4@gmail.com/projectBricks/Governament/Csv's/dataset_DEFAULT_IMF.STA_MFS_CBS_24.0.0.csv"

df_spark = spark.read.csv(caminho, header=True, inferSchema=True)

df_spark.createOrReplaceTempView("governament")
```

```python
df_spark = spark.sql(" select * from governament ")
```

```python
df_spark.columns
```

```python
primeira_linha = df_spark.limit(1).toPandas().iloc[0]

monthly_cols = [col for col in df_spark.columns if '-M' in col]

dates = []
values = []

for col in monthly_cols:
    val = primeira_linha[col]
    if pd.notna(val) and val != '':
        try:
            dates.append(col)
            values.append(float(val))
        except:
            pass
plt.figure(figsize=(16, 6))
plt.plot(dates, values, marker='o', linewidth=2, markersize=4)

step = 24
indices = range(0, len(dates), step)
plt.xticks(indices, [dates[i] for i in indices], rotation=45, ha='right')

plt.xlabel('Período', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.title(f"{primeira_linha['COUNTRY']} - {primeira_linha['INDICATOR']}", 
          fontsize=14, fontweight='bold', pad=22)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print(f"Total de pontos plotados: {len(values)}")
```
