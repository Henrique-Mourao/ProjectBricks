```python
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import os



caminho = "/Workspace/Users/henriquegmour4@gmail.com/projectBricks/Biblioteca/Csv's/minha-biblioteca_2024.csv"

df_spark = spark.read.csv(caminho, header=True, inferSchema=True)

df_spark.createOrReplaceTempView("biblioteca_table")
```

```python
biblioteca = spark.sql("SELECT * FROM biblioteca_table")

print(biblioteca.columns)
display(biblioteca.limit(5))
```

```python
display(biblioteca.groupBy("Tipo de Acesso").count())
```

```python
biblioteca.head
```

```python
df_plot = biblioteca.groupBy(" Mês").sum("Sessões de Estudo").orderBy(" Mês").toPandas()

fig, ax = plt.subplots(figsize=(20, 6))
fig.set_tight_layout(True)
plt.fill_between(df_plot[" Mês"], df_plot["sum(Sessões de Estudo)"], alpha=0.3, color='#2789FF')
plt.xlim(1, 12) 
plt.title('Evolução de Sessões de Estudo por Mês', fontsize=14, fontweight='bold')
plt.xlabel('Mês')
plt.ylabel('Total de Sessões')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
